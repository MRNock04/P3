```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from numpy import transpose as tr
```

```python
# reading the data
HeNe = tr(np.genfromtxt('Data/HeNe.csv', delimiter='\t'))
TiSaph = tr(np.genfromtxt('Data/TiSaph.csv', delimiter='\t'))
Verdi = tr(np.genfromtxt('Data/verdi.csv', delimiter='\t'))
```

```python
# plotting HeNe and Verdi
plt.bar(HeNe[0],HeNe[1], color='orangered', label = 'HeNe')
plt.bar(Verdi[0],Verdi[1], color='limegreen', label = 'Verdi')

# making the plot look nice
plt.title('Spektren des Nd:YAG und des HeNe')
plt.xlabel('Wellenl√§nge [nm]')
plt.ylabel('Helligkeit')
plt.legend()
plt.show()
```

```python
# plotting TiSaph
plt.bar(TiSaph[0],TiSaph[1], color = 'crimson')

# making the plot look nice
plt.title('Fluoreszenzspektrum des Titan-Saphir-Kristalls')
plt.xlabel('Wellenl√§nge [nm]')
plt.ylabel('Helligkeit')
plt.show()
```
